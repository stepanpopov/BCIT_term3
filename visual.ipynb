```python
import requests
URL = 'http://127.0.0.1:5000'
def make_url(cnt):
    res = URL + '/num/' + str(cnt)
    return res
def get_bin_coefs(cnt):
    url = make_url(cnt)
    r = requests.get(url)
    return r.json()
    
```

```python
# Биномиальные коэффициенты от n = 0 до 5
bin_coefs_from_1_to_5 = get_bin_coefs(5)
bin_coefs_from_1_to_5
```

```python
# Биномиальные коэффициенты от n = 0 до 10
bin_coefs_from_1_to_10 = get_bin_coefs(10)
bin_coefs_from_1_to_10
```

```python
# Построение графика для определенного n
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker

def plot_show(data):
    y = data
    x = list(range(0, len(y)))

    fig, ax = plt.subplots()
    ax.bar(x, y, color = 'r')

    ax.set_yticks(data)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.set_xlabel('Степень x в разложении')
    ax.set_title('Коэффициенты бинома Ньютона для n = {}'.format(len(data) - 1))

    fig.set_size_inches(10, 10)
    plt.show()
```

```python
plot_show(bin_coefs_from_1_to_10[-1])
```
